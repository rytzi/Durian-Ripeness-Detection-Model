```python
import numpy as np
import seaborn as sns
import sys
import pandas as pd
import matplotlib.pyplot as plt
import wave
from scipy.io.wavfile import write
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from scipy.signal import butter, lfilter
from sklearn.preprocessing import StandardScaler
import pickle
from sklearn.model_selection import KFold, cross_val_score
from tqdm.notebook import tqdm_notebook
from matplotlib.colors import ListedColormap
from matplotlib.colors import LinearSegmentedColormap
```

```python
confusion_matrix = np.array([[3, 7],
                             [5, 5]])

class_labels = ["Ripe", "Unripe"]

colors = [(0, "#FFFFFF"),
          (1, "#00FF00")]

custom_cmap = LinearSegmentedColormap.from_list("custom_gradient", colors)

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, cmap=custom_cmap, fmt="d", xticklabels=class_labels, yticklabels=class_labels, alpha=0.5)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for CNN")

plt.show()
```
